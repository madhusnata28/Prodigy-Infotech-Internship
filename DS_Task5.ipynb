```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/MADHUSNATA KAR/Documents/Prodigy_Infotech/Task5_sheet.csv")
data
```

```python
print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
```

```python
data.info
```

```python
data.describe()
```

```python
sns.countplot(x='Day_of_Week',data=data)
plt.title('Frequency of accidents with respect to each day')
plt.show()

#Here 1 represents Sunday, 2 for Monday, 3 for Tuesday, 4 for Wednesday, 5 for Thursday, 6 for Friday, 7 for Saturday 
```

```python
sns.barplot(x='Number_of_Casualties', y='Accident_Severity',data=data)
plt.title('No. of casualties with respect to accident severity')
plt.show()
```

```python
sns.countplot(x='Road_Surface_Conditions',data=data)
plt.title('Frequency of accidents with respect to road condition')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_sev=data.groupby('Road_Surface_Conditions')['Accident_Severity'].value_counts()
print("Average severity of accidents by road conditions:")
print(avg_sev)
```

```python
avg_sev=data.groupby('Light_Conditions')['Accident_Severity'].value_counts()
print("Average severity of accidents by light conditions:")
print(avg_sev)
```
